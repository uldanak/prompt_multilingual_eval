```python
from datasets import load_dataset
dataset = load_dataset("glue", "sst2")
```

```python
print(dataset)
```

```python
print(dataset["train"].column_names)
print(dataset["train"][0])
print("train:", len(dataset["train"]))
print("validation:", len(dataset["validation"]))
print("test:", len(dataset["test"]))
```

```python
import pandas as pd

val_df = pd.DataFrame(dataset["validation"])
val_df.to_csv("sst2_val_en_full.csv", index=False)
print("Saved:", val_df.shape)
```

```python
import random

train_pool = dataset["train"].shuffle(seed=42).select(range(5000))
train_pool_df = pd.DataFrame(train_pool)
train_pool_df.to_csv("sst2_train_en_pool_5k.csv", index=False)
print("Saved:", train_pool_df.shape)
```
